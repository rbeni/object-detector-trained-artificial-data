```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
path = '../models/yolo/real_images/'
```

```python
files = os.listdir(path)
files
```

```python
image = cv2.imread(os.path.join(path, files[0]), cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.show()
```

```python
roi = image[500:6000, 1000:6500]
```

```python
plt.imshow(roi, cmap='gray')
plt.show()
```

```python
for i, filename in enumerate(files):
    image = cv2.imread(os.path.join(path, filename), cv2.IMREAD_GRAYSCALE)
    roi = image[500:6000, 1000:6500]
    roi = cv2.resize(roi, (500, 500))
    cv2.imwrite(os.path.join(path, 'p'+filename.replace('.jpg', '.png')), roi)
```
